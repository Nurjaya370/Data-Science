## Control Flow
- Merupakan sebuah bentuk/struktur yang digunakan sebagai pengendali/kontrol dari sebuah bahasa pemrograman
- dalam python terdapat control flow, yaitu penyataan kondisional, dan perulangan.
#### 1. Pernyataan Kondisional
- Pernyataan kondisional itu di kenal juga dengan istilah percabangan. Percabangan pada bahasa pemrograman python menggunakan perintah if,  Else ,Elif. 
- Struktur percabangan mengevaluasi beberapa ekspresi, yang menghasilkan kondisi TRUE atau FALSE saat aplikasi dijalankan.

#### 1.1. If Tunggal
- if tunggal digunakan untuk mengeksekusi statement jika kondisi bernilai benar, Jika kondisi bernilai salah maka tidak ada statement yang di-eksekusi.
- Sintaks dasar If Tunggal
if condition1:statements1
statement default



```python
# contoh if tunggal 1
angka= int(input('masukan angka ='))
if (angka<9):
    print(f'nilai {angka} lebih kecil dari 9')
print('Bye....Bye.....!')
```

```python
# contoh if tunggal 2 (menggunakan operator aritmatika)
if(angka%2==0):
    print(f'nilai {angka} adalah bilangan genap')
print(f'nilai {angka} adalah bilangan ganjil')
```

#### 2. If Else
- if else merupakan Pernyataan kondisi yang dikenal sebagai bentuk dari if sederhana. Pernyataan if dipadukan dengan else
- bentuk dasar sintaks if-else
if condition1:
   statements1
else: 
   statements2
statement default


```python
# contoh if else
z=13
x= int(input('masukan nilai X ='))
if (x==z):
    print(f'nilai {x} sama dengan {z}')
else:
    print(f'nilai {x} tidak sama dengan {z}')
print('Bye....Bye.....!')
```

#### 1.3.if elif
- Bentuk if elif disebut pula kondisi berantai atau seleksi kondisi banyak arah dikarenakan ada beberapa seleksi terhadap kondisi.
- Keyword elif merupakan singkatan dari else-if yang digunakan untuk memeriksa lebih dari satu kondisi. 
if (kondisi 1):
    statement1
elif (kondisi2):
    statement2
elif (kondisi3):
    statement3
else:
    statement4
statement default

```python
x = int(input("\nmasukkan nilai mahasiswa = "))

if (x >= 80 and x <=100):
    print(f"Nilai Mahasiswa = {x} Kategori Grade A ")
elif (x>= 70):
    print(f"Nilai Mahasiswa = {x} Kategori Grade B ")
elif (x >= 50):
    print(f"Nilai Mahasiswa = {x} Kategori Grade C ")
elif (x >= 35):
    print(f"Nilai Mahasiswa = {x} Kategori Grade D ")
elif(x > 35 and x < 0):
    print(f"Nilai Mahasiswa = {x} Kategori Grade E ")
else:
    print(f"Nilai anda masukan tidak sesuai ")
print("\nterima kasih")
```

#### If Bersarang (Nested If)
- if bersarang digunakan untuk melakukan Seleksi terhadap suatu kondisi yang didalamnya terdapat pula kondisi lainnya (if didalam if).
- sintaks dasar Penulisan
- if (kondisi 1):


```python
#contoh if bersarang
x=int(input('masukan angka ='))
if (x>=10):
    if (x%2==0):
        print(f'angka {x} adalah bilangan puluhan dan juga bilangan genap')
    else:
        print(f'angka {x} adalah bilangan puluhan dan juga bilangan ganjil')
else:
    if (x%2==0):
        print(f'angka {x} adalah bilangan satuan dan juga bilangan genap')
    else:
        print(f'angka {x} adalah bilangan satuan dan juga bilangan ganjil')
print('BYE......BYE.....!')
```

## 2	Perulangan (Looping)

- Pernyataan perulangan memungkinkan kita untuk mengeksekusi pernyataan atau sekelompok pernyataan beberapa kali. 
- Python mengenal beberapa pernyataan untuk mengontrol perulangan atau iterasi. 
1.	While
- Digunakan untuk mengulangi pernyataan saat kondisi yang diberikan adalah TRUE.
- menguji kondisi sebelum mengeksekusi badan pengulangan.
2.	For
- Digunakan untuk mengeksekusi urutan pernyataan beberapa kali dan menyingkat kode yang mengelola variabel perulangan.
3.	Nested Perulangan (Pengulangan Bersarang)
- Anda dapat menggunakan satu atau lebih pengulangan di dalam bentuk pengulangan while, atau pengulangan for.

#### 2.1	While
- perulangan while akan memeriksa kondisi pertama dan jika itu benar, maka kontrol akan bergerak di dalam perulangan
- menjalankan pernyataan di dalam perulangan sampai kondisi menjadi salah
- perulangan while, digunakan ini ketika kita tidak yakin berapa kali perlu mengeksekusi sekelompok pernyataan 
- atau dikatakan ketika kita tidak yakin tentang jumlah iterasi.


## Flowchart Perulangan While


```python
# Contoh perulangan WHile
a=10
i=1
n_awal=0
while(i<=a):
    print(f"{i} lebih kecil dari sama dengan {a}")
    n_while=n_awal+i
    i=i+1
else:print(f"pengulangan terjadi sebanyak {n_while} kali") 
```

```python
# Contoh perulangan WHile
sisi=int(input("\nmasukan nilai sisi ="))
print('\n')
x=sisi
jumlah=1
while (jumlah <= sisi):
        print(" " * jumlah,str(jumlah)*x)
        x -=1
        jumlah +=1
print("\nkeluar dari looping")
```

- step adalah jarak antar angka yang dihasilkan oleh range.

```python
print(18*"#"+" PARAMETER STAR "+18*"#")
for i in range(6):
     print(f"perulangan ke {i}")
```

# fungsi Break
- Fungsi break dalam perulangan dapat digunakan untuk mengubah aliran di dalam suatu perulangan.
- menghentikan iterasi perulangan yang sedang berjalan dan mengeksekusi pernyataan berikutnya diluar badan perulangan tersebut. 

```python
# contoh perulangan (while) dengan fungsi break
a=int(input("Masukkan nilai a:"))

i=1
n_awal=0
while(i<=a):
    print(f"{i} lebih kecil dari atau sama dengan {a}")
    n_while=n_awal+i
    i=i+1
    if i==4:
        break
    print("belajar looping dengan while")
print(f"pengulangan terjadi sebanyak {n_while} kali") 
```

```python
# contoh perulangan (for) dengan fungsi break
for perulangan in 'Python': 
    if perulangan == 'h':
        break
    print ('karakter saat ini :', perulangan)
print('keluar darinlooping dengan fungsi BREAK')
```

# Fungsi Continue
Berbeda dengan fungsi break, fungsi continue berguna untuk menghentikan iterasi yang sedang berlangsung dan mentransfer kendali atau control ke iterasi berikutnya di dalam perulangan tersebut, jadi tidak sampai keluar dari badan perulangan.

```python
# Contoh perulangan WHile
sisi=int(input("\nmasukan nilai sisi ="))
print('\n')
jumlah=1
spasi=(sisi/2)
space=int(spasi)
while True:
    if(jumlah%2):
        print(" " * space,"#"*jumlah)
        space -= 1
        jumlah +=1
    else:
        jumlah +=1
        continue
    if jumlah > sisi:
        break
print("\nkeluar dari looping")
```

```python
# contoh perulangan (for) dengan fungsi continue
for perulangan in 'Python': 
    if perulangan == 'h':
        continue
    print ('karakter saat ini :', perulangan)
print('keluar dari looping')
```

# Fungsi Pass
- Penggunaan Fungsi Pass merupakan sebuah sintaksis saja, tetapi tidak ingin statement atau perintah tersebut di eksekusi atau di jalankan.

```python
# contoh perulangan (for) dengan fungsi pass
for i in range(10): 
    if i%2==0:
        pass 
    else:
        print (i)
print('keluar dari looping')
```

```python
# contoh perulangan (for) dengan fungsi pass
for i in range(10): 
    if i%2!=0:
        print (i)
print('keluar dari looping')
```
