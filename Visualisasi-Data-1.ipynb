# Visualisasi Data (Data Visualization)
- Visualisasi data adalah proses menggunakan elemen visual seperti diagram, grafik, atau peta untuk merepresentasikan data (https://aws.amazon.com), Dengan kata lain, ini adalah proses mengubah data hasil analisis menjadi gambaran visual berupa bagan, peta, grafik, dan banyak lainnya.
- Visualisasi data merupakan cara terbaik untuk menyajikan data kepada audiens non-teknis supaya mereka tidak kesulitan untuk membaca dan memahami data tersebut. Ini juga juga memungkinkan para pengambil keputusan dalam memahami konsep sulit atau mengidentifikasi pola baru.

# Peran Visualisasi Data
1. Mengidentifikasi Pola dan Tren.
- visualisasi mampu mengidentifikasi pola dan tren yang tidak bisa ditemukan sebelumnya, dalam dunia bisnis, mengidentifikasi tren sangat diperlukan untuk menentukan strategi bisnis apa yang bisa diterapkan di masa depan.
- Dengan memperhatikan pola-pola yang ada, perusahaan dapat fokus ke area tertentu yang sesuai dengan kebutuhan bisnis.
2. Membuat Data Lebih Mudah Dipahami
- Tujuan adanya visualisasi data adalah menerjemahkan data itu ke bentuk yang lebih mudah dipahami semua orang. Dengan visualisasi data kamu dapat membantu orang melihat dan memahami data kompleks, terlepas dari tingkat keahlian mereka.
3. Memberi Informasi dengan Cepat
- Visualisasi data yang “merangkum” keseluruhan cerita dari data, akan mempercepat dan memudahkan pemberian informasi, tanpa harus mencari informasi itu secara manual.
4. Meningkatkan Bisnis
- Visualisasi data menghasilkan wawasan yang kredibel dan akurat. Saat para eksekutif, stakeholder, manajer, dan jajaran tim lainnya memahami data dengan baik, mereka bisa menindaklanjuti data untuk dijadikan strategi bisnis.

# Jenis Visualisasi Data


1. Table
- Tabel memiliki kolom dan baris. Kolom adalah sumbu Y dan baris adalah sumbu X. dibandingkan jenis visualisasi data lainnya, tabel lebih efektif untuk menampilkan data kualitatif secara terorganisir dan menjelaskan hubungan antar datanya.
2. Graph
- Grafikberupa diagram garis, kurva, titik, atau area yang kehadirannya mewakili variabel tertentu. Grafik biasanya digambarkan sepanjang dua sumbu pada sudut siku-siku.
3. Chart
- Chart adalah visualisasi data dalam bentuk peta, diagram, grafik, atau tabel. Data yang ditampilkan ditandai dengan dua sumbu.
- Banyak jenis chart yang bisa dipilih sesuai kebutuhan data, seperti pie charts, line graph, bar chart, stacked bar graph, dan banyak lagi.

4. Dashboard
- Dashboard adalah tampilan visual dari kumpulan data yang tersedia. Banyaknya data itu akan ditampilkan dalam satu bagian atau file guna membantu menyajikan dan menganalisis data.
5. Geospatial
- Geospatial adalah visualisasi data dalam bentuk peta. Ini umumnya menggunakan warna dan bentuk berbeda untuk menunjukkan perbedaan dan hubungan antara satu potongan data dengan lainnya.
- Sebagai contoh, dalam peta terdapat sungai, gunung, jalan raya, dan banyak lainnya. Masing-masing diberi warna berbeda untuk membedakan satu dengan lainnya.
6. Infographic
- Kombinasi antara visual dan kata-kata. Infografis biasanya disajikan dengan diagram atau bagan. Infografis merupakan salah satu cara efektif untuk menyajikan data dan menjelaskan masalah kompleks dengan cara efisien dan waktu yang cepat.

```python
# Cek data Outlier
# Dengan Menggunakan Inter Quartil Rate
q1 = df["tenure"].quantile(0.25)
q3 = df["tenure"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr

df.loc[(df["tenure"] < fence_low) | (df["tenure"] > fence_high)]
```

### Visualisasi Multivariat
- Visualisasi data multivariat adalah teknik untuk memvisualisasikan data yang terdiri dari tiga atau lebih variabel.
- Tujuannya adalah untuk memahami hubungan antara variabel-variabel tersebut secara simultan. 
- Teknik ini dapat membantu dalam menganalisis dan memahami pola dan korelasi antara beberapa variabel.

```python
#### Menggunakan Scatterplot matrix
- Scatter plot matrix menampilkan beberapa scatter plot di mana setiap plot menampilkan hubungan antara dua variabel dalam dataset
- sementara diagonal menampilkan distribusi dari setiap variabel.
```

```python
sns.pairplot(df, hue='Churn', diag_kind='hist')
```

- Hasil visualisasi diatas akan menampilkan hubungan antara setiap pasangan variabel dalam dataset, serta distribusi dari setiap variabel.
- Garis yang berbeda menunjukkan pemisahan antara pelanggan yang churn (Churn=Yes) dan pelanggan yang tidak churn (Churn=No).

#### Menggunakan Heatmap
- Heatmap dapat membantu dalam memvisualisasikan pola korelasi antara variabel-variabel dalam dataset, yang dapat memberikan wawasan penting dalam analisis data.

```python
# Membuat korelasi matriks untuk semua variabel numerik
corr_matrix = df.corr()

# Membuat heatmap
sns.heatmap(corr_matrix, cmap='coolwarm')

plt.show()
```

- Hasil visualisasi diatas akan menampilkan heatmap yang menunjukkan korelasi antara semua variabel numerik dalam dataset.
- Semakin dekat dengan warna merah, semakin tinggi korelasi positif antara dua variabel.
- Sebaliknya, semakin dekat dengan warna biru, semakin tinggi korelasi negatif antara dua variabel.
