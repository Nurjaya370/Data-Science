## Tipe data Sets (Himpunan)

- Menemukan perbedaan antara dua himpunan
- Union : menggabungkan set, dan hanya menyimpan elemen unik
- Persimpangan : elemen mana yang ada di kedua set
- Menemukan himpunan bagian dan superset

## Tipe data Mapping (Dictionary)
- Tipe data ini bersifat tidak terurut dan mutable (dapat diterapkan perubahan nilai). 
- Tipe data dictionary (Kamus) berisi item yang tidak dapat diurutkan secara lengkap.
- Tipe dictionary berisi item yang terdiri dari pasangan kunci (key) dan nilai, dimana penulisan setiap kunci dipisahkan dari nilainya oleh titik dua (:)
- Sedang setiap item nilai dipisahkan dengan koma, dan semuanya diapit oleh kurung kurawal.
- Dalam dictionary, Kunci (key) merupakan tipe data unik yang nilainya tidak mungkin diberubah
- Sedangkan item nilai dapat berupa tipe apa pun. 

```python
# Membuat Tipe Data Dictionary
kamuskosong={}
kamus1={'k1':"kunci 1",'k2':"kunci2",'k3':123}
print(5*"#"+"TIPE DATA DICTIONARY "+5*"#"+"\n")
print("membuat sebuah kamus kosong ="+ "\n",(kamuskosong))
print("ini adalah bentuk dasar penulisan kamus ="+"\n",(kamus1))
print("ini adalah kunci pada kamus1 ="+"\n",(kamus1.keys()))
print(f"ini adalah item nilai pada kamus1 ="+"\n",(kamus1.values()))
```

```python
dic1={"nama":"elzio","usia":'6 tahun','alamat':'tangerang'}
dic2={1:"Raffanda",2:'Elzio',3:'Jaya'}
print(10*"#"+"MENGAKSES DICTIONARY"+10*"#")
print("ini adalah contoh tipe data dictionary"+"\n",(dic1))
print((dic1['nama']))
print((dic1['usia']))
print((dic1['alamat']))
print(dic2[1])
```

## Operator Bitwise 
- Operator bitwise beroperasi pada satu atau lebih pola bit pada level bit (binary digit: 0 dan 1). 
1. &: Bitwise AND
2. |: Bitwise OR
3. ~: Bitwise NOT membalikan nilai bits dari sisi fositif ke sisi negatif
4. ^: Bitwise XOR
5. >>: Bitwise Geser Kanan
6. <<: Bitwise Geser Kiri

## Operator keanggotaan (membership)
- digunakan untuk mencari keberadaan suatu nilai atau variabel di dalam sequence tertentu baik berupa string, list, tuple, maupun dictionary.
- Operator ini terdiri dari 2 operator, yaitu
1. In : akan bernilai true jika nilai operand sebelah kiri merupakan (anggota), dan terdapat didalam struktur data.
2. Not In : akan bernilai true jika nilai operand sebelah kiri bukan merupakan (anggota), dan tidak terdapat didalam struktur data.

```python
x_list=[1, 3, 5, 7, 9]
y_tuple=('N', 'E', 'T', 'I', 'N', 'D', 'O')
print('\n######## Operator in #######')
print('bernilai TRUE jika sebuah nilai terdapat didalam list')
print( 'nilai 5 ada didalam X list = ',5 in x_list)
print( 'nilai 12 ada didalam X list = ', 12 in x_list)
print('\n######## Operator not in #######')
print('bernilai TRUE jika sebuah nilai tidak terdapat didalam list')
print( 'huruf A tidak ada didalam Y tuple = ', 'A' not in y_tuple)
print( 'huruf N tidak ada didalam Y tuple = ', 'N'  not in y_tuple)
```

## Operator identitas
- berfungsi untuk membandingkan identitas memori dengan nilai sebuah objek, ada dua operator identitas, yaitu.
1. Is: Memeriksa apakah nilai di sebelah kiri operan memiliki identitas memori yang sama dengan nilai di sebelah kanan operan. Jika sama maka kondisi bernilai True
2. Not Is: Memeriksa apakah nilai di sebelah kiri operan memiliki identitas memori yang berbeda dengan nilai di sebelah kanan operan. Jika berbeda maka kondisi bernilai True

```python
a=6
b=8
c=6
print('nilai a = ',a)
print('nilai b = ',b)
print('nilai c = ',c)
print("############# Operator is #############\n")
print('bernilai TRUE jika nilai kedua variabel sama')
print('nilai a = b : ', a is b)
print('nilai a = c : ', a is c)
print('tipe data variabel a adalah integer : ', type(a) is int)
print("\n############# Operator is not #############\n")
print('bernilai TRUE jika nilai kedua variabel berbeda')
print('nilai a berbeda dengan nilai b  = ', a is not b)
print('nilai a berbeda dengan nilai c= ', a is not c)
print('tipe data variabel a bukan integer : ', type(a) is not int)
```
