```python
#membuat dataframe 2 dimensi dengan tipe data list
df1=pd.DataFrame([["k01","Deni","Jakarta",13],
                ["k02","Ani","BSD",45],
                ["k03","Jeni","Pamulang",20],
                ["k04","Alam","Tangerang",11],
                ["k05","Lala","Tangsel",3]],
               columns=["Id","Nama","Alamat","Pembelian"])
df1
```

# Concat
- Digunakan untuk menggabungkan/merekatkan/menempelkan/menambahkan, satu dataframe kepada dataframe lain
- fungsi concat() akan membuat salinan lengkap dari data (append).
- Saat menggabungkan DataFrames, pandas akan berusaha mempertahankan nama indeks/kolom.


### Menggabungkan Dataframe Ke Bagian Bawah
- jika kedua dataframe yang ingin digabungkan memiliki feature/kolom yang sama


```python
df_concat0 = pd.concat([df1, df2])
df_concat0
```

```python
df_concat1 = pd.concat([df1, df2], ignore_index=True)
df_concat1
```

#### Menggabungkan Dataframe Pada Bagian Kanan
- Dilakukan ketika ingin menggabungkan 2 buah dataframe, yang memiliki feature/kolom yang berbeda


```python
df_concat2 = pd.concat([df1, df3], axis=1)
df_concat2
```

```python
df_concat3 = pd.concat([df1, df3.drop(columns="Kd")], axis=1)
df_concat3
```

# Merge
- menggabungkan data berdasarkan pada 1 kriteria atau lebih
- terdapat beberapa bentuk pnggabungan data, yaitu:
1. left joint
2. right joint
3. inner joint
4. full joint


```python
data3 = pd.DataFrame({
    "Id": ["K01", "K02", "K03", "K04", "K05"],
    "Nama": ["Budi", "Jeni", "Bobi", "Tania", "Barli"],
    "Pembelian": [10,15,20,35,4],
    "Kota": ["Bandung", "Jakarta", "Bogor", "Tangerang", "Bali"]
    })
data3
```

```python
data4 = pd.DataFrame({
    "Kd": ["K01", "K02", "K05", "K06", "K07", "K08", "K09"],
    "Pembayaran": [2400,3100,2300,4000,3500,2800,3400]
    })
data4
```

```python
# Merge= menggabungkan 2tabel, default= inner join
data3.merge(data4, left_on= "Id", right_on = "Kd")
```

#### 1. Left joint
- menggabungkan dua buah table, dengan mempertahankan/mengutamakan data/nilai pada tabel sebelah kiri

- 

```python
# Left Join
data3.merge(data4, how = "left", left_on = "Id", right_on = "Kd")
```

#### 1. Right Joint
- menggabungkan dua buah tabel, dengan mempertahankan/mengutamakan data/nilai pada tabel sebelah kanan
- 

```python
# Left Join
data3.merge(data4, how = "right", left_on = "Id", right_on = "Kd")
```

#### 3. Inner Joint
- menggabungkan dua buah tabel, berdasarkan kepada irisan data/nilai dari kedua tabel
- 

```python
# Inner Joint
data3.merge(data4, how = "inner", left_on = "Id", right_on = "Kd")
```

#### 4. Outer/Full Joint
- menggabungkan dua buah tabel secara keseluruhan
- 

```python
# Outter Joint (Full Joint)
data3.merge(data4, how = "outer", left_on = "Id", right_on = "Kd")
```

```python
# Outter Joint (Full Joint)
data3.merge(data4, how = "outer", left_on = "Id", right_on = "Kd").drop(columns="Kd")
```

# Pivot
- tenik atau untuk menghitung, meringkas, dan menganalisis data yang memungkinkan Anda melihat perbandingan, pola, dan tren dalam data Anda
- Fungsi Pandas Pivot menghasilkan tabel pivot berdasarkan 3 atau lebih kolom dari dataFrame (indeks, kolom, dan nilai).
- (index) Indeks: sebagai sumbu pada baris (parameter pengelompokan)
- (Columns) Kolom: sebagai parameter pengelompokan kolom
- (values) Nilai: nilai agregasi yang ingin kita amati

```python
#pd.isna(dataNSL).sum()
dataNSL=dataNSL.fillna(0)
dataNSL.head()
```

```python
dataNSL.pivot_table(index =['label'], columns ='protocol_type', 
                    values =["duration","src_bytes"], 
                    aggfunc = "max", fill_value=0)
```

```python
dataNSL.pivot_table(index =['label'], columns ='protocol_type', 
                    values =["duration","wrong_fragment"], 
                    aggfunc = "mean", fill_value = 0)
```

```python
dataNSL.pivot_table(index =['label'], columns ='protocol_type', 
                    values =["duration","wrong_fragment"], 
                    aggfunc = len, fill_value = 0)
```

#### 1. Apply
- fungsi apply diterapkan didalam tingkat series dan dataframe multidimensi

```python
df0 = pd.Series([60,50,30,40],
                index=['Adam', 'Dani', 'Icha', 'Haris'])
df0
```

```python
# Fungsi apply pada series
tambah=df0.apply(lambda x:x + 5)
tambah
```
