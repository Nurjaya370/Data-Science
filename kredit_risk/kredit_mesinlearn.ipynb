```python
#contoh operasi aritmatika
a = np.array([1, 2, 3, 4, 5, 6])
b = np.array([6, 5, 4, 3, 2, 1])
c = a + b
print(f"array 1 = ",a,"\n")
print(f"array 2 =",b,"\n")
print(f"hasil operasi penjumlahan array 1 dan 2 =",c,"\n")
# contoh oerasi statistik
rerata = np.mean(a)  # Rata-rata
stddev = np.std(a)  # Deviasi standar
print(f"contoh operasi statistik pada array",rerata,"\n")
# contoh manipulasi array
reshaped = c.reshape((2, 3))  # Mengubah bentuk array
con = np.concatenate((a, b))  # Menggabungkan dua array
split_array=np.split(con,3)
print(f"contoh manipulasi pada array","\n",reshaped)
print(f"contoh manipulasi array 2",con,"\n")
print(f"contoh manipulasi array 3","\n",split_array)
```

```python
import matplotlib.pyplot as plt  
import numpy as np  

x = np.arange(1, 6)  
y1 = np.array([1, 2, 3, 4, 5])  
y2 = np.array([2, 3, 4, 5, 6])  

plt.fill_between(x, y1, color="blue", alpha=0.4)  
plt.fill_between(x, y2, color="salmon", alpha=0.5)  
plt.title('Area Plot Contoh')  
plt.xlabel('X-axis')  
plt.ylabel('Y-axis')  
plt.show()
```

```python
import seaborn as sns  

# Mengambil dataset yang tersedia di web/ server seaborn   
print(sns.get_dataset_names()) 
```

```python
# Menggunakan dataset bawaan
dataDF=sns.load_dataset('titanic')
dataDF
```

```python
dataDF.info()
```

```python
plt.figure(figsize=(10, 6))  

# Membuat histogram  
hist_plot = sns.histplot(dataDF['age'], bins=30, kde=False, color='lime', stat='density', alpha=0.5)  

# Menambahkan garis KDE berwarna biru  
sns.kdeplot(dataDF['age'], color='blue', linewidth=2)  

# Menambahkan pengaturan plot  
plt.title('Distribusi Usia Penumpang')  
plt.xlabel('Usia')  
plt.ylabel('Frekuensi')  
  

# Menampilkan plot  
plt.show()
```

```python
import matplotlib.pyplot as plt  
import seaborn as sns  


plt.figure(figsize=(10, 6))  

# Membuat histogram  
hist_plot = sns.histplot(dataDF['age'], bins=30, kde=False, color='lime', stat='density', alpha=0.5)  

# Menambahkan garis KDE berwarna biru  
sns.kdeplot(dataDF['age'], color='blue', linewidth=2)  

# Menambahkan pengaturan plot  
plt.title('Distribusi Usia Penumpang')  
plt.xlabel('Usia')  
plt.ylabel('Frekuensi')  


# Menampilkan plot  
plt.show()
```

```python
sns.boxplot(data=dataDF, x='class', y='age', palette='Set2')
plt.title('Box Plot Contoh')
plt.show()
```

```python
# Menghitung korelasi
correlation = dataDF.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Contoh')
plt.show()
```

```python
sns.pairplot(dataDF[['class', 'age', 'sex', 'alone']], hue='alone')  
plt.title('Pair Plot Contoh')  
plt.show()
```

```python
sns.violinplot(data=dataDF, x='class', y='fare', hue='sex', split=True)
plt.title('Violin Plot Contoh')
plt.show()
```
