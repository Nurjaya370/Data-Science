## Mengaplikasikan sebuah fungsi kedalam sebuah tabel pada pandas DataFrame
- Dengan menggunakan fungsi lambda
- Dengan menggunakan np.vectorize

### Dengan menggunakan fungsi lambda
- digunakan untuk memasukan data secara perbaris. np.vectorize digunakan untuk memasukan data dalam satu kolom

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'Nama':('andi','kaka','nana','dimas','lala'),
    'Nilai':(40,60,60,70,90)
})
df
```

```python
def grading(nilai):
    if nilai >=80 and nilai < 100:
        return 'Grade A'
    elif nilai >=75:
        return 'Grade B'
    elif nilai >=60:
        return 'Grade C'
    elif nilai >=35:
        return 'Grade D'
    else:
        return 'Grade E'
```

```python
%%timeit
df['Grade_lambda']=df.apply(lambda x: grading(x['Nilai']), axis=1)
```

```python
%%timeit
df['Grade_numpy'] = np.vectorize(grading)(df['Nilai'].values)
```
